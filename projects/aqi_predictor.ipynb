```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# 3 plots per row
cols_per_row = 3
num_cols = len(numeric_cols)
rows = (num_cols + cols_per_row - 1) // cols_per_row  # calculate rows

fig, axes = plt.subplots(rows, cols_per_row, figsize=(15, 4 * rows))
axes = axes.flatten()  # flatten axes array for easy indexing

for i, col in enumerate(numeric_cols):
    sns.boxplot(x=df[col], color='skyblue', ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}', fontsize=12)
    axes[i].set_xlabel(col)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
X=df.drop(['Country','City','AQI_Category','CO_AQI_Category','Ozone_AQI_Category','NO2_AQI_Category','PM2.5_AQI_Category','AQI_Value'],axis=1)
```

```python
X
```

```python
y=df['AQI_Value']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=44)

```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()

lr_model.fit(X_train,y_train)
```

```python
lr_model.score(X_test,y_test)
y_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
r2score=r2_score(y_test,y_pred)
msc=mean_squared_error(y_test,y_pred)
print("r2 score : ",r2score)
print('mean squared error :',msc)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

L1 REGULARIZATION (LASSO)

```python
from sklearn.linear_model import Lasso,Ridge
lasso_model=Lasso()
lasso_model.fit(X_train,y_train)
```

```python
lasso_model.score(X_test,y_test)
```

L2 REGULARIZATION (RIDGE)

```python
from sklearn.linear_model import Ridge
ridge_model=Ridge(alpha=0.5)
ridge_model.fit(X_train,y_train)
ridge_model.score(X_test,y_test)

y_pred=ridge_model.predict(X_test)
```

```python
ridge_model.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2score = r2_score(y_test, y_pred)

print('R Score:', r2score)
print('Mean Squared Error:', mse)
```
